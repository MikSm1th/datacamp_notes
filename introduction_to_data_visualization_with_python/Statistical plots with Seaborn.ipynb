```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn import datasets
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set(rc={'figure.figsize':(11,8)})
```

### Visualizing Regressions

```python
tips = pd.read_csv('tips.csv', usecols=[1,2,3,4,5,6,7])
tips.head()
```

```python
tips = sns.load_dataset('tips')
sns.lmplot(x='total_bill', y='tip', data=tips);
```

```python
sns.lmplot(x='total_bill', y='tip', hue='sex', palette='Set1', data=tips);
```

```python
sns.lmplot(x='total_bill', y='tip', col='sex', data=tips);
```

```python
sns.residplot(x='total_bill', y='tip', data=tips, color='indianred');
```
